```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style and seed for reproducibility
sns.set(style="whitegrid")
np.random.seed(42)

# Generate different distributions
normal_data = np.random.normal(50, 10, 200)  # Normal/Gaussian
skewed_data = np.random.exponential(10, 200) + 30  # Right-skewed
uniform_data = np.random.uniform(30, 70, 200)  # Uniform
bimodal_data = np.concatenate([
    np.random.normal(35, 5, 100),
    np.random.normal(65, 5, 100)
])  # Bimodal

# Combine data for plotting
all_data = [normal_data, skewed_data, uniform_data, bimodal_data]
labels = ['Normal', 'Right-Skewed', 'Uniform', 'Bimodal']

# Create comparison plots
plt.figure(figsize=(16, 12))
```

```python
# 1. Box Plot Comparison
plt.subplot(1, 1, 1)
sns.boxplot(data=all_data, orient='h', palette='Set3')
plt.yticks(range(len(labels)), labels)
plt.title('Box Plot Comparison')
plt.xlabel('Value')
```

```python
# 2. Violin Plot Comparison
plt.subplot(1, 1, 1)
sns.violinplot(data=all_data, orient='h', palette='Set3')
plt.yticks(range(len(labels)), labels)
plt.title('Violin Plot Comparison')
plt.xlabel('Value')
```

```python
# 3. Histogram Comparison (overlay with transparency)
plt.subplot(1, 1, 1)
colors = ['blue', 'green', 'red', 'purple']
for i, (data, label, color) in enumerate(zip(all_data, labels, colors)):
    sns.kdeplot(data, label=label, color=color, fill=True, alpha=0.3)
plt.title('Density Plot Comparison')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
```

```python
# 4. Histogram grid (small multiples)
plt.subplot(1, 1, 1)
for i, (data, label) in enumerate(zip(all_data, labels)):
    plt.hist(data, bins=20, alpha=0.7, density=True,
             histtype='step', linewidth=2, label=label)
plt.title('Histogram Comparison (Normalized)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Print statistics for comparison
print("Statistical Comparison of Distributions:")
print("-" * 60)
print(f"{'Distribution':<15} {'Mean':<10} {'Median':<10} {'Std Dev':<10} {'Skewness':<10}")
print("-" * 60)

from scipy.stats import skew

for data, label in zip(all_data, labels):
    mean = np.mean(data)
    median = np.median(data)
    std_dev = np.std(data)
    skewness = skew(data)
    print(f"{label:<15} {mean:<10.2f} {median:<10.2f} {std_dev:<10.2f} {skewness:<10.2f}")
```
